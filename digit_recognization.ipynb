```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Load Dataset
from sklearn.datasets import load_digits
dataset=load_digits()
dataset
```

```python

#Summerize the Data
print(dataset.data)
print(dataset.target)
print(dataset.images.shape)
print(dataset.data.shape)
images_length=len(dataset.images)
print(images_length)
```

```python
#Sample Output
plt.matshow(dataset.images[5])
plt.gray()
plt.show()
```

```python
#Segregatting the Data
x=dataset.images.reshape((images_length,-1))
y=dataset.target
```

```python
#Splitting the Data Train and Test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=35)
```

```python
#Creating the Model
from sklearn import svm
model=svm.SVC()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
#Predicting the Model
n=76
result=model.predict(dataset.images[n].reshape((1,-1))) 
plt.matshow(dataset.images[n],cmap=plt.cm.gray_r,interpolation='nearest')
plt.title(result)
print(result)
plt.show()
```

# New Section

```python

##Evaluting the Model
from sklearn.metrics import accuracy_score
print("accuracy of model:{}%".format(accuracy_score(y_test,y_pred)*100))
```
